```python
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
```

```python
subfig = make_subplots(specs=[[{"secondary_y": True}]])
```

```python
url = 'https://raw.githubusercontent.com/pbelousova/371-Lab/main/371_Lab3_Heater_Pot_Therm.csv'
df1 = pd.read_csv(url)
```

```python
fig1 = px.line(df1, x = 'Relative_Time_[ms]', y = 'Potentiometer_Value_[AU]')
fig1.show()
```

```python
url = 'https://raw.githubusercontent.com/pbelousova/371-Lab/main/371_Lab3_Heater_Therm.csv'
df2 = pd.read_csv(url)
```

```python
fig2 = px.line(df2, x = 'Relative_Time_[ms]', y = 'Thermistor_Value_[AU]')
fig2.show()
```

```python
fig2.update_traces(yaxis="y2")
```

```python
subfig.add_traces(fig1.data + fig2.data)
subfig.layout.xaxis.title="Relative_Time_[ms]"

subfig.layout.yaxis.title="Potentiometer_Value_[AU]"
subfig.layout.yaxis.color='blue'

subfig.layout.yaxis2.title="Thermistor_Value_[AU]"
subfig.layout.yaxis2.color='red'

subfig.for_each_trace(lambda t: t.update(line=dict(color=t.marker.color)))

subfig.show()
```
