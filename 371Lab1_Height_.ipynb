```python
import pandas as pd
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/pbelousova/371-Lab/main/Pressure%20%5BhPa%5D.csv'
df1 = pd.read_csv(url)
```

```python
fig = px.line(df1, x = 'Relative time [ms]', y = 'Pressure [hPa]', title='Pressure vs. Time')
fig.show()
```

```python
import numpy as np

# T = 288.16 K @ sea-level for C
P_ref = 101325
C = -8619.222


url = 'https://raw.githubusercontent.com/pbelousova/371-Lab/main/Pressure%20%5BhPa%5D.csv'
df1 = pd.read_csv(url)

df1 ['Height [m]'] = C*np.log(100* df1['Pressure [hPa]']/ P_ref )

fig = px.line(df1, x = 'Relative time [ms]', y = 'Height [m]', title='Height vs. Time')
fig.show()
```
